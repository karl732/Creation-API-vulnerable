```python
from faker import Faker
import random
from datetime import date

fake = Faker("fr_FR")

roles = ["Customer", "Customer", "Customer", "Analyst"]
countries = ["FR", "BE", "CH", "CA"]

n_users = 50
start_id = 5

rows = []

for i in range(n_users):
    
    user_id = start_id + i
    full_name = fake.name()
    username = full_name.lower().replace(" ", ".")
    password = fake.password(length=6)
    role = random.choice(roles)
    country = random.choice(countries)
    signup_date = fake.date_between(start_date="-2y", end_date="today")

    row = f"({user_id}, '{username}', '{password}', '{full_name}', '{role}', '{country}', '{signup_date}')"
    
    rows.append(row)

print("INSERT INTO dim_user")
print("(`user_id`,`username`,`password`,`full_name`,`user_role`,`country`,`signup_date`)")
print("VALUES")

print(",\n".join(rows) + ";")
```

```python
fake = Faker()

account_types = ["Savings", "Checking", "Business"]
currencies = ["EUR", "USD", "GBP"]
statuses = ["Active", "Active", "Active", "Suspended", "Closed"]

n_accounts = 50
start_id = 4

rows = []

for i in range(n_accounts):

    account_id = start_id + i
    account_type = random.choice(account_types)
    currency = random.choice(currencies)
    open_date = fake.date_between(start_date="-3y", end_date="today")
    status = random.choice(statuses)
    
    # solde réaliste
    balance = round(random.uniform(-200, 50000), 2)

    row = f"({account_id}, '{account_type}', '{currency}', '{open_date}', '{status}', {balance})"
    rows.append(row)

print("INSERT INTO dim_account")
print("(`account_id`, `account_type`, `currency`, `open_date`, `status`, `balance`)")
print("VALUES")
print(",\n".join(rows) + ";")
```

```python
from faker import Faker
import random

fake = Faker()

n_transactions = 100
start_id = 5

# IDs existants dans tes tables
user_ids = list(range(1, 51))      # 50 users
account_ids = list(range(1, 51))   # 50 comptes

transaction_types = ["Deposit", "Withdrawal", "Transfer"]

rows = []

for i in range(n_transactions):

    transaction_id = start_id + i
    user_id = random.choice(user_ids)
    account_id = random.choice(account_ids)

    transaction_date = fake.date_time_between(start_date="-1y", end_date="now")

    transaction_type = random.choice(transaction_types)

    amount = round(random.uniform(10, 2000), 2)

    # signe du montant selon type
    if transaction_type == "Deposit":
        amount = abs(amount)

    else:  # Withdrawal ou Transfer
        amount = -abs(amount)

    row = f"({transaction_id}, {user_id}, {account_id}, '{transaction_date}', '{transaction_type}', {amount})"
    rows.append(row)

print("INSERT INTO fact_transactions")
print("(`transaction_id`,`user_id`,`account_id`,`transaction_date`,`transaction_type`,`amount`)")
print("VALUES")
print(",\n".join(rows) + ";")
```
